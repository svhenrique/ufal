```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

## 3)

```python
pontos = np.random.standard_normal(size=10000) # gerando 10000 pontos de distribuição normal padrão
pontos
```

```python
pontos2 = np.random.standard_normal(size=10000) # gerando 10000 pontos de distribuição normal padrão  com um deslocamento de 4

valores = []
# for para incrementar os valores em 4 e construir deslocamento
for ponto in pontos2:
  valores.append(ponto+4)
```

```python
pontos_concatenados = np.concatenate([pontos, valores])
pontos_concatenados
```

```python
plt.hist(pontos_concatenados, bins=100, color="white", ec="black")
plt.title("Mixed distribution histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

## 5)

```python
dados5 = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/up_down_expression.txt', delimiter="\t")
dados5.head()
```

```python
# separando os dados de acordo com a classificação State
unchanging = dados5.loc[dados5['State']=='unchanging']
up = dados5.loc[dados5['State']=='up']
down = dados5.loc[dados5['State']=='down']
```

```python
plt.scatter(unchanging.Condition1, unchanging.Condition2, color="gray")
plt.scatter(up.Condition1, up.Condition2, color="red")
plt.scatter(down.Condition1, down.Condition2, color="blue")
plt.xlabel("Condition1")
plt.ylabel("Condition2")
plt.show()
```
